```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['Age_mean','age']]
```

```python
df['Age_median'] = df['age'].fillna(df['age'].median())
```

```python
df[['Age_mean', 'Age_median', 'age']]
```

```python
df['embarked'].isnull().sum()
```

```python
df['mode_embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df[['mode_embarked', 'embarked']]
```

```python
df['mode_embarked'].isnull().sum()
```

```python
df = df.dropna(subset=['Age_mean', 'Age_median', 'mode_embarked'],inplace=)
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='mean')
```

```python
df2['age'] = si.fit_transform(df2[['age']])
```

```python
df2['age'].isnull().sum()
```

```python
si2 = SimpleImputer(strategy='most_frequent')
```

```python
df2.head(2)
```

```python
df2['embarked'] = si2.fit_transform(df2[['embarked']])[:, 0]
```
